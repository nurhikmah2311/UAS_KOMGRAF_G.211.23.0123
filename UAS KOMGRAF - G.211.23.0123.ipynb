a. Distribusi data pada setiap atribut Purchase Price, Current Price dan Gain/Loss dengan grafik histogram


```python
fig, axes = plt.subplots(3, 1, figsize=(10, 15))
colors = sns.color_palette("Set2", 3)
sns.histplot(df['Purchase Price'], bins=30, kde=True, ax=axes[0], color=colors[0], alpha=0.7)
axes[0].set_title('Distribusi Purchase Price')
axes[0].set_xlabel('Purchase Price')
axes[0].set_ylabel('Frekuensi')

sns.histplot(df['Current Price'], bins=30, kde=True, ax=axes[1], color=colors[1], alpha=0.7)
axes[1].set_title('Distribusi Current Price')
axes[1].set_xlabel('Current Price')
axes[1].set_ylabel('Frekuensi')

sns.histplot(df['Gain/Loss'], bins=30, kde=True, ax=axes[2], color='red')
axes[2].set_title('Distribusi Gain/Loss')
axes[2].set_xlabel('Gain/Loss')
axes[2].set_ylabel('Frekuensi')
```

b. Atribut Stock Name dan Symbol yang memiliki Dividen / Share 5 tertingi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca data
df = pd.read_csv("/content/stock.csv")

# Mengkonversi Dividend/Share jika diperlukan
df["Dividend/share"] = df["Dividend/share"].str.lstrip("$").astype(float)

# Filter data dengan dividen 5 tertinggi
top_5_dividends = df.nlargest(5, "Dividend/share")

# Menggabungkan simbol dan stock name
top_5_dividends["Symbol and Stock Name"] = top_5_dividends["Symbol"] + " - " + top_5_dividends["Stock Name"]

# Visualisasi menggunakan seaborn
custom_palette = sns.color_palette("coolwarm", len(top_5_dividends))
sns.set_style("whitegrid")
plt.figure(figsize=(12, 6))
sns.barplot(x="Symbol and Stock Name", y="Dividend/share", hue="Symbol and Stock Name", data=top_5_dividends, palette=custom_palette, dodge=False)
plt.xlabel("Symbol dan Stock Name")
plt.ylabel("Dividend / Share")
plt.title("5 Stock Name dan Symbol dengan Dividen/Share 5 tertinggi")
plt.xticks(rotation=45)
plt.legend([],[], frameon=False)
plt.tight_layout()
plt.show()
```

c. Atribut Stock Name dan Symbol yang memiliki Market Value 5 tertinggi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv("/content/stock.csv")

# Menghapus tanda dolar
df["Market Value"] = df["Market Value"].str.replace("$", "")

# Menghapus tanda koma (pemisah ribuan) dan mengubah titik (pemisah desimal) menjadi koma
df["Market Value"] = df["Market Value"].str.replace(",", "").str.replace(".", "")

# Konversi string menjadi float
df["Market Value"] = df["Market Value"].astype(float)

# Filter 5 nilai Market Value tertinggi
top_5_market_value = df.nlargest(5, "Market Value")
top_5_market_value["Symbol and Stock Name"] = top_5_market_value["Symbol"] + " - " + top_5_market_value["Stock Name"]

# Visualisasi menggunakan seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x="Symbol and Stock Name", y="Market Value", hue="Symbol and Stock Name", data=top_5_market_value, palette="viridis")
plt.legend([],[], frameon=False)
plt.xlabel("Symbol dan Stocok Name")
plt.ylabel("Market Value")
plt.title("5 Stock Name dan Symbol dengan Market Value 5 tertinggi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

d. Atribut Stock Name dan Symbol yang memiliki Annual Yield 7 tertingi

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read data
df = pd.read_csv("/content/stock.csv")

# Menghapus tanda persen dan mengonversinya ke float
df["Annual Yield"] = df["Annual Yield"].str.replace("%", "").astype(float)

# Filter 7 nilai Annual Yield tertinggi
top_7_annual_yield = df.nlargest(7, "Annual Yield")

top_7_annual_yield["Symbol and Stock Name"] = top_7_annual_yield["Symbol"] + " - " + top_7_annual_yield["Stock Name"]

# Visualisasi menggunakan seaborn
plt.figure(figsize=(12, 6))
sns.set(style="whitegrid")
sns.barplot(x="Symbol and Stock Name", y="Annual Yield", hue="Symbol and Stock Name", data=top_7_annual_yield, palette="plasma", dodge=False)
plt.legend([],[], frameon=False)
plt.xlabel("Symbol dan Stocok Name")
plt.ylabel("Annual Yield")
plt.title("5 Stock Name dan Symbol dengan Annual Yield 7 tertinggi")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
